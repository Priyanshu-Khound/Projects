# Infant Mortality Rate Analysis

### Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 

warnings.filterwarnings('ignore')
```

### Reading the data

```python
data=pd.read_excel("Infant-mortality-rate_2020.xlsx")
```

```python
data
```

#### Droping NaN values

```python
data=data.dropna(axis=1)
data=data.reset_index(drop=True)
```

```python
data
```

```python
data.isna().sum()
```

```python
data['Country.Name'].nunique()
```

### Dropping Upper and Lower Values

```python
df_median=data[data['Uncertainty.Bounds*']=='Median']
```

```python
df_median
```

## Visualizations

### Visulizations to check best and worst performing countries in 1985-1986

#### Bar Graph to show Infant Mortality Rate in the year 1985-1986

```python
df_median_sorted=df_median.sort_values(by=['1985.5'])
d=pd.concat([df_median_sorted.head(10), df_median_sorted.tail(10)],axis=0)

fig, ax=plt.subplots(figsize=(23,10))
sns.barplot(d['1985.5'],d['Country.Name'],ax=ax)
ax.set_title('Best and Worst Performing Countries in 2009-2010')
ax.set_xlabel('IMF as of 1985-1986')
```

#### Time Series Graph to Show the best and the worst performing Countries in the year 1985-1986 and their improvement in recent years

```python
arr=d.drop(['ISO.Code','Country.Name','Uncertainty.Bounds*'],axis=1)
n1=np.array(arr[0:1])
n1=n1.flatten()
n2=np.array(arr[1:2])
n2=n2.flatten()
n3=np.array(arr[2:3])
n3=n3.flatten()
n4=np.array(arr[3:4])
n4=n4.flatten()
n5=np.array(arr[4:5])
n5=n5.flatten()
n6=np.array(arr[5:6])
n6=n6.flatten()
n7=np.array(arr[6:7])
n7=n7.flatten()
n8=np.array(arr[7:8])
n8=n8.flatten()
n9=np.array(arr[8:9])
n9=n9.flatten()
n10=np.array(arr[9:10])
n10=n10.flatten()

L=[n1,n2,n3,n4,n5,n6,n7,n8,n9,n10]
fig,ax=plt.subplots(figsize=(24,10))
x_axis=np.array(d.drop(['ISO.Code','Country.Name','Uncertainty.Bounds*'],axis=1).columns)

for i in L:
    sns.lineplot(x_axis,i,ax=ax)

ax.legend(labels=d['Country.Name'].head(10),fontsize=20)
```

In the best performing countries in the year 1985-1986, the best improvements can be seen in Norway, Iceland and Monaco.

## Conclusion

From the analysis of above data we are able to deduce that most the best performing countries are developed countries, and are mostly european countries, with very good health infrastructure and facilities. 

While most of the worst performing countries are under-developed African countries. With very poor to non health infrastructure, poor living standards, and many dieases infecting many people every year. These countries were mostly exploited for their resources, leaving them with poor conditions.

For improving infant mortaility rate in poor performing countries, the health infrastructure in those countries should be improved, with more investment in healthcare and research, this is the only way to save life, it is clear from the data of best performing countries.
